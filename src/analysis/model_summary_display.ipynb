```python
import pandas as pd
```

```python
df = pd.read_csv('../../assets/summarised_data/model_performance.csv')
df = df.set_index('Model')
df['Generalisation Gap'] = df['Best train F1'] - df['Best test F1']
df = df.sort_values('Best test F1')
```

```python
styler = df.drop(columns='Generalisation Gap')
styler = styler.style.background_gradient(cmap='Greens',
                                      subset=['Best train F1',
                                              'Best validation set F1',
                                              'Best test F1'],)



styler = styler.format(precision=3)

```
